# City school analysis

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("D:/Data analytics bootcamp/week4/module_4_challenge/panda-challenge/Starter_Code (1)/Starter_Code/PyCitySchools/Resources/schools_complete.csv")
student_data_to_load = Path("D:/Data analytics bootcamp/week4/module_4_challenge/panda-challenge/Starter_Code (1)/Starter_Code/PyCitySchools/Resources/students_complete.csv")

# read in the dataset:
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

#merging the two datasets
merged_data = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
merged_data.head()
```

# District Summary

```python
# Total Number of Unique Schools
total_schools = len(merged_data['School ID'].unique())

# Total Number of Students
total_students = len(merged_data['Student ID'].unique())
total_students_fmt = '{:,}'.format(total_students) # format with comma separators

# Total Budget
total_budget = school_data['budget'].sum()
total_budget_fmt = '${:,.0f}'.format(total_budget) # format with dollar sign symbol

# Average Math Score
avg_math_score = student_data['math_score'].mean()

# Average Reading Score 
avg_reading_score = student_data['reading_score'].mean()

# % Passing Math
passing_math = student_data[student_data['math_score'] >= 70].count()['Student ID']/total_students * 100

# % Passing Reading
passing_reading = student_data[student_data['reading_score'] >= 70].count()['Student ID']/total_students * 100

# Overall Passing Rate
overall_passing = merged_data[(merged_data['math_score'] >= 70) & (merged_data['reading_score'] >= 70)].count()['Student ID']/total_students * 100
```

```python
#creating a DataFrame to display the district summary

district_summary = pd.DataFrame({
    "Total Schools": [total_schools],
    "Total Students": [total_students_fmt],
    "Total Budget": [total_budget_fmt],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [passing_math],
    "% Passing Reading": [passing_reading],
    "% Overall Passing": [overall_passing]
})

district_summary
```

# School summary

```python
# Calculate the total number of students per school
student_count = merged_data.groupby(['school_name']).size()

# Calculate the total budget per school
budget_per_school = merged_data.groupby(['school_name'])['budget'].mean()

# Calculate the budget per student per school
budget_per_student = budget_per_school / student_count

# Calculate the average math score per school
average_math_score = merged_data.groupby(['school_name'])['math_score'].mean()

# Calculate the average reading score per school
average_reading_score = merged_data.groupby(['school_name'])['reading_score'].mean()

# Calculate the percentage of students who passed math per school
passing_math_count = merged_data[merged_data['math_score'] >= 70].groupby(['school_name']).size()
percent_passing_math = passing_math_count / student_count * 100

# Calculate the percentage of students who passed reading per school
passing_reading_count = merged_data[merged_data['reading_score'] >= 70].groupby(['school_name']).size()
percent_passing_reading = passing_reading_count / student_count * 100

# Calculate the overall passing percentage per school
overall_passing_percent = (percent_passing_math + percent_passing_reading) / 2
```

```python
# Combine all the above calculations into a summary DataFrame
summary_df = pd.DataFrame({
    'School Type': school_data.set_index('school_name')['type'],
    'Total Students': student_count,
    'Total School Budget': budget_per_school,
    'Per Student Budget': budget_per_student,
    'Average Math Score': average_math_score,
    'Average Reading Score': average_reading_score,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    '% Overall Passing Rate': overall_passing_percent
})
```

```python
# Format the columns appropriately
summary_df['Total Students'] = summary_df['Total Students'].map('{:,}'.format)
summary_df['Total School Budget']=summary_df['Total School Budget']. map('${:,.2f}'.format)
summary_df['Per Student Budget']=summary_df['Per Student Budget']. map('${:.2f}'.format)

# Display the summary DataFrame
summary_df
```

# Math Scores by Grade

```python
# Create a pivot table to display the average math score for students of each grade level at each school
math_scores_by_grade = pd.pivot_table(student_data, values = 'math_score', index=['school_name'], columns = ['grade'], aggfunc = 'mean')

# Update the column names to include "Math"
math_scores_by_grade.columns = ['9th Grade Math', '10th Grade Math', '11th Grade Math', '12th Grade Math']

# Remove the index name
math_scores_by_grade.index.name = None

# Display the math scores by grade DataFrame
math_scores_by_grade
```

# Reading Scores by Grade


```python
# Create a pivot table to display the average reading score for students of each grade level at each school
reading_scores_by_grade = pd.pivot_table(student_data, values = 'reading_score', index=['school_name'], columns = ['grade'], aggfunc = 'mean')

# Update the column names to include "Reading"
reading_scores_by_grade.columns = ['9th Grade Reading', '10th Grade Reading', '11th Grade Reading', '12th Grade Reading']

# Remove the index name
reading_scores_by_grade.index.name = None

# Display the reading scores by grade DataFrame
reading_scores_by_grade
```

# Scores by School Spending

```python
summary_df.info()
```

```python
# Create four bins with reasonable cutoff values to group school spending
spending_bins = [0, 585, 630, 645, 680]

# Create labels for these bins
labels = ['$585', '$585-630', '$630-645', '$645-680']

# Use pd.cut to categorize spending based on the bins
summary_df['Spending Ranges (Per Student)'] = pd.cut(summary_df['Per Student Budget'], bins=spending_bins, labels=labels)
```

```python
# Calculate mean scores per spending range
spending_math_scores = summary_df.groupby(['Spending Ranges (Per Student)'])['Average Math Score'].mean()
spending_reading_scores = summary_df.groupby(['Spending Ranges (Per Student)'])['Average Reading Score'].mean()
spending_passing_math = summary_df.groupby(['Spending Ranges (Per Student)'])['% Passing Math'].mean()
spending_passing_reading = summary_df.groupby(['Spending Ranges (Per Student)'])['% Passing Reading'].mean()
overall_passing_spending = summary_df.groupby(['Spending Ranges (Per Student)'])['% Overall Passing Rate'].mean()
```

```python
# Create a DataFrame to summarize scores by school spending
spending_summary = pd.DataFrame({
    'Average Math Score': spending_math_scores,
    'Average Reading Score': spending_reading_scores,
    '% Passing Math': spending_passing_math,
    '% Passing Reading': spending_passing_reading,
    '% Overall Passing Rate': overall_passing_spending
})
```

```python
# Format the columns appropriately
spending_summary['Average Math Score'] = spending_summary['Average Math Score'].map('{:.2f}'.format)
spending_summary['Average Reading Score'] = spending_summary['Average Reading Score'].map('{:.2f}'.format)
spending_summary['% Passing Math'] = spending_summary['% Passing Math'].map('{:.2f}'.format)
spending_summary['% Passing Reading'] = spending_summary['% Passing Reading'].map('{:.2f}'.format)
spending_summary['% Overall Passing Rate'] = spending_summary['% Overall Passing Rate'].map('{:.2f}'.format)

# Display the spending summary DataFrame
spending_summary
```
